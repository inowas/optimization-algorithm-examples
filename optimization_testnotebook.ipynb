```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Here some basic test calculation of sin, cos, tan
x_vals = np.arange(-10, 10, 0.5)
y_vals = {"sin": np.sin(x_vals),
          "cos": np.cos(x_vals),
          "tan": np.tan(x_vals)}
```

```python
# A plot of the values
# sine
plt.plot(x_vals, y_vals["sin"], color="red", label="sine")
# cosine
plt.plot(x_vals, y_vals["cos"], color="blue", label="cosine")
# tangent
plt.plot(x_vals, y_vals["tan"], color="green", label="tangent")
# legend
plt.legend()
# grid
plt.grid()
```
